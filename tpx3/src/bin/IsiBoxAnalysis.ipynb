```python
%matplotlib qt
import hyperspy.api as hs
import numpy
import matplotlib.pyplot as plt
import csv
plt.rc('font', size=18) #controls default text size
```

```python
file = numpy.fromfile("isi_raw105_output.isi", dtype='int32')
unique, counts = numpy.unique(file, return_counts=True)
```

```python
ss = 256
frames = 2300
channels = 4
si = numpy.zeros(ss * ss * frames * channels, dtype='int16')
si[unique] += counts
#si = numpy.reshape(si, (channels, frames, ss, ss))
si = numpy.reshape(si, (frames, ss, ss, channels))
si = numpy.roll(si, 100, axis=1)
```

```python
s = hs.signals.Signal1D(si)
s = s.rebin(scale=[1, 1, 1, 1])
```

```python
for x in range(4):
    if (s.isig[x].sum().data[0] > 0):
        print(f'Plotting channel {x}')
        s.isig[x].transpose(navigation_axes=1).plot()
        s.isig[x].transpose(navigation_axes=1).T.plot()
```

### Maps

```python
file1 = numpy.fromfile("isi_g2.txt", dtype='int64')
file2 = numpy.fromfile("isi_g2_index.txt", dtype='int32')
```

```python
index = file2 * 999 + (file1 + 499)
unique, counts = numpy.unique(index, return_counts=True)
```

```python
ss = 256
bins = 999
si = numpy.zeros(ss * ss * bins, dtype='int16')
si[unique] += counts
si = numpy.reshape(si, (ss, ss, bins))
si = numpy.roll(si, 100, axis=0)
```
